```python
data["18_group"] = data["18_group"].fillna(value = 0)
data["is_mi"] = data["is_mi"].fillna(value = 0)
data["17_group"] = data["17_group"].fillna(value = 0)
data["blended"] = data["blended"].fillna(value = "")
data["is_ml_student"] = data["is_ml_student"].fillna(value = "")
data.head(5)
```

```python
values1 = data.loc[:, "is_first_time"]
values2 = data.loc[:, "id"]
a = []
count = 0

for i, j in zip(values1, values2):
    if i == "Нет" and not(j in a):
        count += 1
        a.append(j)    
count   
```

```python
d = data.loc[:, "percentile"]
a = set()
for i in d:
    a.add(i)

data.shape[0] - len(a)
```

```python
d = data.loc[:, "blended"]
a = set()
for i in d:
    if (a != ''):
        a.add(i)
b = list(a)
b.remove("")
print(b)
print()


c = [0, 0, 0]
for i in d:
    if (i == b[0]):
        c[0] += 1
    elif (i == b[1]):
        c[1] += 1
    elif (i == b[2]):
        c[2] += 1

max = c[0]
indexMax = 0
for i in range(3):
    if max < c[i]:
        max = c[i]
        indexMax = i
print(b[indexMax])        

dr = data.loc[:, "rating"]
maxDevops = 0
maxCompetitionAnalysis = 0
maxInterseption = 0

for i, j in zip(d, dr):
    if i == b[0]:
        if maxDevops < j:
            maxDevops = j
    if i == b[1]:
        if maxCompetitionAnalysis < j:
            maxCompetitionAnalysis = j
    if i == b[2]:
        if maxInterseption < j:
            maxInterseption = j

print(maxDevops, " ", maxCompetitionAnalysis, " ", maxInterseption)            

if (maxDevops > maxCompetitionAnalysis):
    if (maxDevops > maxInterseption):
        print(b[0], " ", c[0])
    else:
        print(b[2], " ", c[2])    
elif (maxCompetitionAnalysis > maxInterseption):
    print(b[1], " ", c[1])
else:
    print(b[2], " ", c[2])    
```

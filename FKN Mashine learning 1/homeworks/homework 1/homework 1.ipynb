В данном случае набор курсов задаётся упорядоченным множеством ('fall_1', 'fall_2', 'fall_3', 'spring_1', 'spring_2', 'spring_3', 'blended'). Элемент 
blended будет нулевым для 3 курса.

```python
temp = pd.DataFrame(data).drop_duplicates(subset=["id"], keep='last')
temp = temp[["fall_1", "fall_2", "fall_3", "spring_1", "spring_2", "spring_3", "blended"]]
temp = temp.groupby(temp.columns.to_list(), as_index=False).size()
temp = temp[temp["size"] >= 2]
temp.drop_duplicates(subset=["fall_1", "fall_2", "fall_3", "spring_1", "spring_2", "spring_3", "blended"])
```

Задание 7:

1) Найдите курсы по выбору, на которые записывались как студенты 18-го года набора, так и студенты 17-го года набора.
2) Методом исключения найдите курсы, которые предлагались только студентам 18-го года и только студентам 17-го года.

Решение:

```python
temp = pd.DataFrame(data).drop_duplicates(subset=["id"], keep='last')

temp1 = temp[temp["17_group"] != 0.0]
temp2 = temp[temp["18_group"] != 0.0]

temp1 = temp1[["fall_1", "fall_2", "fall_3", "spring_1", "spring_2", "spring_3", "blended"]]
temp2 = temp2[["fall_1", "fall_2", "fall_3", "spring_1", "spring_2", "spring_3", "blended"]]

temp1 = temp1.drop_duplicates(subset=["fall_1", "fall_2", "fall_3", "spring_1", "spring_2", "spring_3", "blended"])
temp2 = temp2.drop_duplicates(subset=["fall_1", "fall_2", "fall_3", "spring_1", "spring_2", "spring_3", "blended"])

a = set([i for i in temp1['fall_1']])
a.update([i for i in temp1['fall_2']])
a.update([i for i in temp1['fall_3']])
a.update([i for i in temp1['spring_1']])
a.update([i for i in temp1['spring_2']])
a.update([i for i in temp1['spring_3']])
a.update([i for i in temp1['blended']])

b = set([i for i in temp2['fall_1']])
b.update([i for i in temp2['fall_2']])
b.update([i for i in temp2['fall_3']])
b.update([i for i in temp2['spring_1']])
b.update([i for i in temp2['spring_2']])
b.update([i for i in temp2['spring_3']])
b.update([i for i in temp2['blended']])

print("a:")
print(a)
print("b:")
print(b)
print("a and b")
print(a & b)
print("a - b")
print(a.difference(b))
print("b - a")
print(b.difference(a))
```
